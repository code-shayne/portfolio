# An Analysis of Spotify and Youtube Songs

### Introduction

The dataset includes 20,717 songs collected from Spotify. For each song there are 26 points of data describing the content of the song as well as how the song rated on various Youtube and Spotify statistics. In my analysis, I will only use 11 of these columns. These 11 columns are as follows:

* Track: name of track <br>
* Artist: name of artist <br>
* Danceability: describes how suitable a track is for dancing based on a combination of musical elements, the scale is from 0.0(least danceable) to 1.0(most danceable) <br>
* Energy: a scale of 0.0 to 1.0 measuring intensity and activity <br>
* Loudness: overall loudess of a track in decibals <br>
* Valence: a measure from 0.0 to 1.0 describing the musical positiveness of a track <br>
* Tempo: overall tepo of the track in beats per minute <br>
* Duration_ms: duration of the track in miliseconds <br>
* Views: number of views on Youtube <br>
* Likes: number of likes on Youtube <br>
* Stream: number of streams on Spotify <br>

I chose this dataset because I am very interested in music and the components of it. The dataset also has a lot of usable data that can be analyzed in multiple different ways. There is a lot of different ways I can compare the numerous variables and they are all related in interesting manners.

### Statistical Analysis

```python
artist_danceability = df_filtered[['Artist','Danceability','Likes']].groupby('Artist', as_index=False).mean()
artist_danceability.sort_values(by='Danceability', ascending=False).head(10)
```

```python
artist_danceability.plot.scatter(x='Danceability',y='Likes',figsize=(15,10))
```

```python
artist_danceability.corr()
```

I created a Dataframe comparing average danceability score to average like count for each artist. This data was then graphed on a scatter plot. The plot shows that there is not a strong correlation between the danceability score and like counts. Analyzing the correlation between the two variables supports this conclusion because the r value is only 0.184.

```python
df_artist_count = df_filtered.groupby('Artist').agg({'Track':'count','Likes':'mean'})
df_artists_more = df_artist_count[df_artist_count['Track'] > 5]
df_artists_less = df_artist_count[df_artist_count['Track'] <= 5]
df_artist_count.head()
```

```python
df_artist_count.plot.scatter(x='Track',y='Likes',figsize=(15,10))
```

```python
from scipy.stats import ttest_ind
ttest_ind(df_artists_more['Likes'],df_artists_less['Likes'])
```

I seperated the artists within the dataset with more tracks(6-10) and the artists with less tracks(1-5) into two seperate dataframes. In a t-test between the average like counts for the artists in each dataset, the pvalue was 0.076, which is larger than the alpha of 0.05. This means there is not a signifigant difference between the like counts, meaning that having more tracks within the dataset does not correlate to being more popular. In contrast, a scatter plot between the number of tracks an artist has and that artist's average like count shows that there is a correlation between having more tracks and being more well liked. Although, according to the t-test, this correlation is not a strong one.

```python
ttest_ind(df_filtered['Danceability'],df_filtered['Energy'])
```

```python
ttest_ind(df_filtered['Energy'],df_filtered['Valence'])
```

```python
ttest_ind(df_filtered['Danceability'],df_filtered['Valence'])
```

I performed three t-tests comparing three variables: danceability vs. energy, energy vs. valence, and danceability vs. valence. All three of the tests showed large signifigant differences. This is likely because I compared different variables to each other instead of the same variables from different trials. The two variables that were relatively the most related were danceability and energy.
